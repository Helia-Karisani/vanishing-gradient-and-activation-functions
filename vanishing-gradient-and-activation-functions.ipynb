```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid function and its derivatives
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```
